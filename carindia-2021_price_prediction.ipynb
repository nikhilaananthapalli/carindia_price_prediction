```python
data['model_name_new']=data['model_name'].str.extract('(\S{3,})')
data['model_name_new'].head(4)
#.str-allows you to access the values of series as a string
#extract-used to extract in the regular expression as columns
#\s+ - matches on ore more white spaces
#\S{3,}-matches three or more consecutive non-white space characters
#\S- matches one or more consecutive non white-space characters
```

```python
data['model_name_new'].value_counts()
```

```python
df1=data.loc[(data['model_name_new']=='Swift')]
```

```python
df1.head()
```

```python
df2=data.loc[(data['model_name_new']=='Alto')]
df2.head()
```

```python
df3=data.loc[(data['model_name_new']=='Wagon')]
df3.head()
```

```python
df=pd.concat([df1,df2,df3],ignore_index=True)
```

```python
df.groupby('model_name_new').count()
```

```python
df.to_csv('model_sep.csv')
```

```python
df.head()
```

```python
df['maker'].value_counts()
```

```python
df=df.drop(['maker','model_name','city'],axis=1)
```

```python
df=pd.get_dummies(df)
```

```python
x=df.drop(['price (₹)'],axis=1)
y=df['price (₹)']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtrain.head(10)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(xtrain,ytrain)
```

```python
y_pred=reg.predict(xtest)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(ytest,y_pred)
```

```python
x.head()
```

```python
results=pd.DataFrame(columns=['Actual','Predicted'])
results['Actual']=ytest
results['Predicted']=y_pred
```

```python
results=results.reset_index()
results['id']=results.index
```
